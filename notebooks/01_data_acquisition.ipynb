```python
"""
Notebook 01: Data Acquisition
==============================

This notebook handles the initial download of the Kepler Objects of Interest (KOI)
cumulative catalog from the NASA Exoplanet Archive.

Output: data/raw/kepler_koi_cumulative.csv

Author: Mersen-cloud
"""

# Standard library
from pathlib import Path

# Third-party
import pandas as pd
from astroquery.ipac.nexsci.nasa_exoplanet_archive import NasaExoplanetArchive

# Display configuration
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 200)

print("Imports successful ✓")
```

```python
# Define paths
PROJECT_ROOT = Path.cwd().parent  # We're inside notebooks/, so go up one level
DATA_RAW = PROJECT_ROOT / "data" / "raw"
DATA_RAW.mkdir(parents=True, exist_ok=True)

OUTPUT_FILE = DATA_RAW / "kepler_koi_cumulative.csv"

print(f"Project root: {PROJECT_ROOT}")
print(f"Data will be saved to: {OUTPUT_FILE}")
```

```python
# Query the cumulative KOI table from NASA Exoplanet Archive
# This may take 30-60 seconds depending on connection

print("Downloading Kepler KOI cumulative table from NASA Exoplanet Archive...")
print("(This can take 30-60 seconds, please wait)")

koi_table = NasaExoplanetArchive.query_criteria(
    table="cumulative",
    select="*"
)

print(f"\n✓ Download complete")
print(f"  Rows: {len(koi_table)}")
print(f"  Columns: {len(koi_table.colnames)}")
```

```python
# Convert astropy table to pandas DataFrame
df = koi_table.to_pandas()

# Save to CSV
df.to_csv(OUTPUT_FILE, index=False)
print(f"✓ Saved to {OUTPUT_FILE}")
print(f"  File size: {OUTPUT_FILE.stat().st_size / 1024:.1f} KB")

# Show shape
print(f"\nDataFrame shape: {df.shape}")
```
